# Figures

Creating various figures for the paper using data from other files.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
pid_results = pd.read_csv('./data/pid_wind.csv')
north_results = pd.read_csv('./data/n_wind.csv')
cardinal_no10m_results = pd.read_csv('./data/cardinal_wind_no10m.csv')
cardinal_results = pd.read_csv('./data/cardinal_wind.csv')
cardinal_vel_results = pd.read_csv('./data/cardinal_wind_vel.csv')
```

```python
def get_polar_data(results, speed): # converts data into format needed for polar plots
    subset = results[results['Wind'].str.contains(str(speed))]
    subset['theta'] = np.array([1.5708, 4.71239, 0, 3.14159, 2.35619, 3.92699, 0.785398, 5.49779])
    subset = subset.sort_values('theta')
    first_row = subset.iloc[0:1, :]
    subset = pd.concat([subset, first_row], ignore_index=True)

    return subset['theta'].to_numpy(), subset['Mean TTE'].to_numpy()
```

```python
pid_theta, pid_r = get_polar_data(pid_results, 10)
north_theta, north_r = get_polar_data(north_results, 10)
cardinal_theta, cardinal_r = get_polar_data(cardinal_results, 10)
cardinal2_theta, cardinal2_r = get_polar_data(cardinal_no10m_results, 10)
cardinal_vel_theta, cardinal_vel_r = get_polar_data(cardinal_vel_results, 10)

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.set_rscale('log')
ax.plot(pid_theta, pid_r, label='PID Results', alpha=0.5)
ax.plot(north_theta, north_r, label='North Results', alpha=0.5)
ax.plot(cardinal_theta, cardinal_r, label='Cardinal High Results', alpha=0.5)
ax.plot(cardinal2_theta, cardinal2_r, label='Cardinal Low Results', alpha=0.5)
ax.plot(cardinal_vel_theta, cardinal_vel_r, label='Cardinal Vel Results', alpha=0.5)
ax.plot(pid_theta, [5]*len(cardinal_r), label='Safety Corridor', color='black', alpha=0.7)

ax.set_title("Average Deviation for Wind Directions (10 m/s)")
ax.legend(loc='lower left', prop={'size':8})

# plt.savefig('./figures/polar10m')
plt.show()
```
